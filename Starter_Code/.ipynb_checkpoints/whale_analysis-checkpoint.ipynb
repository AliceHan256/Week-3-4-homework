# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
whale_data = Path("Resources/whale_returns.csv")
algo_data = Path("Resources/algo_returns.csv")
sp_tsx_history_data = Path("Resources/sp_tsx_history.csv")
```

```python
whale_df = pd.read_csv(whale_data, index_col = "Date", parse_dates=True, infer_datetime_format=True)
whale_df = whale_df.sort_index()
algo_df = pd.read_csv(algo_data, index_col = "Date", parse_dates=True, infer_datetime_format=True)
algo_df = algo_df.sort_index()
sp_tsx_history_df = pd.read_csv(sp_tsx_history_data, index_col = "Date", parse_dates=True, infer_datetime_format=True)
sp_tsx_history_df = sp_tsx_history_df.sort_index()
```

```python
whale_df.dtypes
algo_df.dtypes
sp_tsx_history_df.dtypes
```

```python
# To make sure the dates are a type of DateTimeIndex.
print(whale_df.index)
print(algo_df.index)
print(sp_tsx_history_df.index)
```

```python
# Reading whale returns
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df_new = whale_df.dropna()
whale_df_new
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_df.head()
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df_new = algo_df.dropna()
algo_df_new
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx_history_df.head()
```

```python
# Check Data Types
sp_tsx_history_df.dtypes
```

```python
# Fix Data Types
sp_tsx_history_df["Close"] = sp_tsx_history_df.Close.str.replace("$", "")
sp_tsx_history_df["Close"] = sp_tsx_history_df.Close.str.replace(",", "")
sp_tsx_history_df
```

```python
sp_tsx_history_df["Close"] = sp_tsx_history_df["Close"].astype("float")
sp_tsx_history_df.dtypes
```

```python
# Calculate Daily Returns
sp_tsx_history_returns = sp_tsx_history_df.pct_change()
sp_tsx_history_returns.head()
#sp_tsx_history_returns
```

```python
# Drop nulls
sp_tsx_history_returns = sp_tsx_history_returns.dropna()
sp_tsx_history_returns
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_history_returns = sp_tsx_history_returns.rename(columns={
    "Close": "S&P 500"
})
sp_tsx_history_returns
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
all_returns = pd.concat([whale_df_new, algo_df_new, sp_tsx_history_returns], axis = "columns", join="inner")
all_returns
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P TSX 60 using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
all_returns_rolling_std = all_returns.rolling(window = 21).std()
# Plot the rolling standard deviation
all_returns_rolling_std.plot(figsize = (25, 10), title = "Rolling Standard Deviation for All Portfolios")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
all_returns_corr = all_returns.corr()
# Display de correlation matrix
all_returns_corr.style.background_gradient(cmap="ocean")
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
tiger_global_cov = all_returns["TIGER GLOBAL MANAGEMENT LLC"].rolling(window = 60).cov(all_returns["S&P 500"])
# Calculate variance of S&P TSX
sp_variance = all_returns["S&P 500"].rolling(window=60).var()
# Computing beta
beta = tiger_global_cov / sp_variance 
# Plot beta trend
beta.plot(figsize = (25, 10), title = "Tiger Global management LLC Beta")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
all_returns.ewm(halflife = 21).std().plot(figsize = (25, 10), title = "Exponentially Weighted Average for All Portfolios")
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (all_returns.mean() * 252) / (all_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind = "bar", title = "Sharpe Ratios for All Portfolios")
```

## Calculate average sharpe ratios for each individual portfolio

```python
# # Calculate average sharpe ratios for Whales portfolios
whales_sharpe_ratio = (whale_df_new.mean() * 252) / (whale_df_new.std() * np.sqrt(252))
whales_sharpe_ratio
```

```python
# # Calculate average sharpe ratios for Algorithmic portfolio
algo_sharpe_ratio = (algo_df_new.mean() * 252) / (algo_df_new.std() * np.sqrt(252))
algo_sharpe_ratio
```

```python
# # Calculate average sharpe ratios for S&P 500
sp_tsx_sharpe_ratio = (sp_tsx_history_returns.mean() * 252) / (sp_tsx_history_returns.std() * np.sqrt(252))
sp_tsx_sharpe_ratio
```

## Calculate average sharpe ratios for each portfolio

```python
# Calculate average sharpe ratios for each individual portfolio
whales_sharpe_ratio_avg = whales_sharpe_ratio.mean()
algo_sharpe_ratio_avg = algo_sharpe_ratio.mean()
sp_tsx_sharpe_ratio_avg = sp_tsx_sharpe_ratio.mean()
print(f"Whales portfolio's average sharpe ratio: {whales_sharpe_ratio_avg}")
print(f"Algorithmic portfolio's average sharpe ratio: {algo_sharpe_ratio_avg}")
print(f"S&P 500 portfolio's average sharpe ratio: {sp_tsx_sharpe_ratio_avg}")
```

## Compare sharpe ratios averages

```python
# Compare sharpe ratios average between Whales portfolio and Algorithmic portfolio
whales_sharpe_ratio_avg > algo_sharpe_ratio_avg
```

```python
# Compare sharpe ratios average between S&P 500 and Algorithmic portfolio
sp_tsx_sharpe_ratio_avg > algo_sharpe_ratio_avg
```

### Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.

From the above calculation, Algorithmic portfolios got the highest average sharpe ratios compared to Whales Portfolios' average sharpe ratio. Therefore, the Algorithmic strategies outperform both the market (S&P TSX 60) and the Whales portfolios.

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

For this demo solution, we fetch data from three companies listes in the S&P TSX 60 index.

* `SHOP` - [Shopify Inc](https://en.wikipedia.org/wiki/Shopify)

* `OTEX` - [Open Text Corporation](https://en.wikipedia.org/wiki/OpenText)

* `L` - [Loblaw Companies Limited](https://en.wikipedia.org/wiki/Loblaw_Companies)

```python
# Reading data from 1st stock
shop_historical_data = Path("Resources/shop_historical.csv")
shop_historical_df = pd.read_csv(shop_historical_data, index_col = "Date", parse_dates=True, infer_datetime_format=True)
shop_historical_df = shop_historical_df.sort_index()
shop_historical_df.head()
```

```python
# Reading data from 2nd stock
otex_historical_data = Path("Resources/otex_historical.csv")
otex_historical_df = pd.read_csv(otex_historical_data, index_col = "Date", parse_dates=True, infer_datetime_format=True)
otex_historical_df = otex_historical_df.sort_index()
otex_historical_df.head()
```

```python
# Reading data from 3rd stock
l_historical_data = Path("Resources/l_historical.csv")
l_historical_df = pd.read_csv(l_historical_data, index_col = "Date", parse_dates=True, infer_datetime_format=True)
l_historical_df = l_historical_df.sort_index()
l_historical_df.head()
```

```python
# Combine all stocks in a single DataFrame
all_my_stocks = pd.concat([shop_historical_df, otex_historical_df, l_historical_df], axis = "rows", join="inner")
all_my_stocks.head()
```

```python
# Reset Date index
all_my_stocks = all_my_stocks.reset_index()
all_my_stocks
```

```python
# Reorganize portfolio data by having a column per symbol
portfolio = all_my_stocks.pivot_table(values = "Close", index = "Date", columns = "Symbol")
portfolio.head()
```

```python
# Calculate daily returns
portfolio_daily_returns = portfolio.pct_change()
# Drop NAs
portfolio_daily_returns = portfolio_daily_returns.dropna().copy()
# Display sample data
portfolio_daily_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
portfolio_daily_returns = portfolio_daily_returns.dot(weights)
# Display sample data
portfolio_daily_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_returns["my_portfolio"] = portfolio_daily_returns
all_returns.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_returns = all_returns.dropna().copy()
all_returns.head()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
all_returns_std = all_returns.std() * np.sqrt(252)
all_returns_std
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
all_returns_roll_std = all_returns.rolling(window=21).std()
# Plot rolling standard deviation
all_returns_roll_std.plot(figsize = (25, 10), title = "Rolling Standard Deviation for All Portfolios")
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
all_returns_corr = all_returns.corr()
all_returns_corr.style.background_gradient(cmap="Blues")
```

### Calculate and Plot the 60-day Rolling Beta for Your Portfolio compared to the S&P 60 TSX

```python
# Calculate and plot Beta
my_portfolio_cov = all_returns["my_portfolio"].rolling(window=60).cov(all_returns["S&P 500"])
my_portfolio_var = all_returns["S&P 500"].rolling(window=60).var()
beta = my_portfolio_cov / my_portfolio_var
beta.plot(figsize = (25, 10), title = "My Portfolio Beta")
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualzied Sharpe Ratios
all_portfolio_sharpe_ratio = (all_returns.mean() * 252) / (all_returns.std() * np.sqrt(252))
all_portfolio_sharpe_ratio
```

```python
# Visualize the sharpe ratios as a bar plot
all_portfolio_sharpe_ratio.plot.bar(title = "All Portfolio's Sharpe Ratio")
```

```python
# Calculate average sharpe ratios for my portfolio
my_portfolio_sharpe_ratio_ave = ((portfolio_daily_returns.mean() * 252) / (portfolio_daily_returns.std() * np.sqrt(252))).mean()

# Calculate average sharpe ratios for Algorithmic portfolio
algo_sharpe_ratio_avg = algo_sharpe_ratio.mean()

# Calculate average sharpe ratios for S&P 500
sp_tsx_sharpe_ratio_avg = sp_tsx_sharpe_ratio.mean()

print(f"My portfolio's average sharpe ratio: {my_portfolio_sharpe_ratio_ave}")
print(f"Algorithmic portfolio's average sharpe ratio: {algo_sharpe_ratio_avg}")
print(f"S&P 500's average sharpe ratio: {sp_tsx_sharpe_ratio_avg}")
```

### How does your portfolio do?
Out of all portfolios, Algo 1 conducted the best performance, reaching a highest sharpe ratio, 1.901461, whereas S&P 500 only reached 0.034651, which is also underperformed than my portfolios with the sharpe ratio of 1.377974. Averagely, my portfolios are overperformed than the Algorithmic portfolios and the S&P 500.
